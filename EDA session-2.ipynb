**create data frame using dictionary**

```python
name=['Ayush','Avinash','Akash','Anvi']
age=[25,30,35,40]
pd.DataFrame(zip(name,age),columns=['Names','Age'])
```

```python
dict1={'Names':['Ayush','Avinash','Akash'],
       'Ages':[25,30,35]}
dict1
```

```python
pd.DataFrame(dict1)
# keys become columns 
# values becomes rows
# no need to provide columns names
```
