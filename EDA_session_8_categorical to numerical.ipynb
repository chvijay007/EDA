**map**
- frist get the unique labels for each coumm
- create a dictionary by providing a values to each label
- for example case status has two labels,we are provideing two values
  - assign 0 for certified

```python
visa_df['case_status'].unique()
```

```python
d={'Denied':1,'certified':0}
d
```

```python
visa_df['case_status']=visa_df['case_status'].map(d)
```

```python
visa_df
```

```python
cols=visa_df.select_dtypes(include='object').columns
for i in cols:
    labels=list(visa_df[i].unique())
    values=[i for i in range(len(labels))]
    d=dict(zip(labels,values))
    visa_df[i]=visa_df[i].map(d)
visa_df
```

**one hot encoder**
- one hot menas if one is ON another one is OFF
- ON represent with 1,and OFF represent with 0
- for example case status has two unique labels
   -denied
   - crtifed
- if we apply one hot encoder 

|case_status|case_status_denied|case_status_certified|
|--------|--|--|
|Denied|1|0|
|certified|0|1|

**advantage**
- in machine learing we have one important property is **independent of varible
- machine learing models expectes varibles  are independent each other
- one hot coder creating new columns, the new coimns are independent each other
- independnt varible menas both

**disadvatge**
- one hot encoder creates new columns equal to number of unique lables
- if number of unique lables incress, new columns also increase
- the more columns leads data is more 
- then data is more the complexity will incress
- complexity incress more computation power requied
- also time also incase
